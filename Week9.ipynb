```python
import re
import nltk
from striprtf.striprtf import rtf_to_text
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from collections import defaultdict
import calendar  
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('wordnet')


lemmatizer = WordNetLemmatizer()

stopwordlist = set([
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your', 
    'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', 
    'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their', 
    'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that', 
    'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being', 
    'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 
    'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 
    'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through', 
    'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down', 
    'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 
    'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any', 'both', 
    'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no', 'nor', 'not', 
    'only', 'own', 'same', 'so', 'than', 'too', 'very', 's', 't', 'can', 'will', 
    'just', 'don', 'should', 'now'
])

def clean_text(text):
    text = re.sub(r'http\S+', '', text)  
    text = re.sub(r'#', '', text)       
    text = re.sub(r'[^A-Za-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()  
    return text


def tokenization(text):
    return word_tokenize(text)

def remove_stopwords(tokens):
    return [token for token in tokens if token.lower() not in stopwordlist]


def lemmatize(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

def preprocess_text(text):
    text = clean_text(text)
    tokens = tokenization(text)
    tokens = remove_stopwords(tokens)
    tokens = lemmatize(tokens)
    return tokens


def extract_articles_and_dates(file_path):
   
    with open(file_path, 'r', encoding='utf-8') as file:
        rtf_content = file.read()
        plain_text = rtf_to_text(rtf_content)
    
    articles = plain_text.split("End of Document")
    
    data = []
    for article in articles:
        date_match = re.search(r"Load-Date:\s*([A-Za-z]+\s+\d{1,2},\s+\d{4})", article)
        if date_match:
            full_date = date_match.group(1)

            article_text = re.sub(r"Load-Date:\s*[A-Za-z]+\s+\d{1,2},\s+\d{4}", "", article).strip()
            preprocessed_tokens = preprocess_text(article_text)
            data.append((full_date, preprocessed_tokens))
    
    return data

def count_word_by_month_year(articles, word):
    word = lemmatizer.lemmatize(word.lower()) 
    word_counts = defaultdict(int)
    article_details = [] 
    
    for idx, (full_date, tokens) in enumerate(articles, start=1):
        date_match = re.search(r"([A-Za-z]+)\s+(\d{1,2}),\s+(\d{4})", full_date)
        if date_match:
            month = date_match.group(1)
            year = date_match.group(3)
            month_year = f"{month} {year}"
            
            word_count = tokens.count(word)
            word_counts[month_year] += word_count
            
            article_details.append((idx, full_date, word_count))
    
    return word_counts, article_details

def create_month_year_histogram(word_counts, word):
    sorted_month_years = sorted(word_counts.keys(), key=lambda x: (x.split()[1], list(calendar.month_name).index(x.split()[0])))
    sorted_frequencies = [word_counts[m] for m in sorted_month_years]

    plt.figure(figsize=(12, 6))
    plt.bar(sorted_month_years, sorted_frequencies, color='blue')
    plt.xlabel("Month and Year")
    plt.ylabel("Frequency")
    plt.title(f"Frequency of the word '{word}' by Month and Year")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    file_path = 'beep.rtf'  
articles_and_dates = extract_articles_and_dates(file_path)

word_to_count = "eurostar"
word_frequencies_by_month_year, article_details = count_word_by_month_year(articles_and_dates, word_to_count)
print(f"Word counts for '{word_to_count}' per article:")
for article in article_details:
    print(f"Article {article[0]} - Date: {article[1]}, Word Count: {article[2]}")
create_month_year_histogram(word_frequencies_by_month_year, word_to_count)
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim import corpora, models
import gensim

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

lemmatizer = WordNetLemmatizer()
stopwordlist = set(stopwords.words('english'))

def clean_text(text):
    """
    Perform basic cleaning on text.
    """
    text = re.sub(r'http\S+', '', text)  
    text = re.sub(r'[^a-zA-Z\s]', '', text) 
    text = re.sub(r'\s+', ' ', text).strip() 
    text = text.lower()  
    return text

def tokenization(text):
    """
    Tokenize text into words.
    """
    return word_tokenize(text)

def remove_stopwords(tokens):
    """
    Remove stopwords from a list of tokens.
    """
    return [token for token in tokens if token not in stopwordlist]

def lemmatize(tokens):
    """
    Lemmatize tokens to their base form.
    """
    return [lemmatizer.lemmatize(token) for token in tokens]

def preprocess_text(text):
    """
    Preprocess text by applying cleaning, tokenization, stopword removal, and lemmatization.
    """
    text = clean_text(text)
    tokens = tokenization(text)
    tokens = remove_stopwords(tokens)
    tokens = lemmatize(tokens)
    return tokens

def split_articles(text, delimiter="End of Document"):
    """
    Split the input text into individual articles based on a delimiter.
    """
    articles = text.split(delimiter)
    return [article.strip() for article in articles if article.strip()]

def extract_title(article):
    """
    Extract the title (first line) of an article.
    """
    lines = article.splitlines()
    title = lines[0].strip() if lines else "Untitled Article"
    return title

def extract_body(article):
    """
    Extract the body of an article, ignoring metadata.
    """
    lines = article.splitlines()
    body = " ".join(lines[1:]).strip() if len(lines) > 1 else ""
    return body

def process_articles(file_path):
    """
    Read a text file, split it into articles, and process each article individually.
    """
    with open(file_path, 'r', encoding='utf-8', errors='replace') as file:
        content = file.read()

    articles = split_articles(content)

    processed_articles = []
    for article in articles:

        title = extract_title(article)
        body = extract_body(article)


        processed_body = preprocess_text(body)

        processed_articles.append({
            "title": title,
            "processed_body": processed_body
        })

    return processed_articles

def topic_modeling(processed_articles, num_topics=10, num_words=10):
    """
    Perform LDA topic modeling on the processed articles and display topics in a readable format.
    """

    processed_texts = [article['processed_body'] for article in processed_articles]


    dictionary = corpora.Dictionary(processed_texts)
    corpus = [dictionary.doc2bow(text) for text in processed_texts]

    lda_model = gensim.models.ldamodel.LdaModel(
        corpus=corpus,
        id2word=dictionary,
        num_topics=num_topics,
        random_state=100,
        update_every=1,
        chunksize=100,
        passes=10,
        alpha='auto',
        per_word_topics=True
    )

    topics = lda_model.print_topics(num_words=num_words)
    print("\nTopics Found:\n")
    for topic_num, topic in topics:

        parsed_words = " ".join([word.split("*")[1].replace('"', '').strip() for word in topic.split("+")])
        print(f"Topic #{topic_num + 1}: {parsed_words}")

if __name__ == "__main__":
    file_path = 'beep.txt' 
    processed_articles = process_articles(file_path)

    print("\nPerforming Topic Modeling...\n")
    topic_modeling(processed_articles, num_topics=10, num_words=10)
```

```python
import ollama
import time
from datetime import datetime
from collections import Counter

PROMPT = """Classify this article into one or more of these categories: 'Policy', 'Railway Updates', 'Eurostar Monopoly', 'Delays', or 'Other'. 
Choose only from these categories. Return the categories in a comma separated list. Do not provide any explanation or additional information."""

def normalize_categories(raw_category):
    """
    Normalize the categories by removing extraneous text and formatting inconsistencies.
    """

    categories = [
        cat.strip("'\" ").strip() 
        for cat in raw_category.split(",")
    ]
    valid_categories = {'Policy', 'Railway Updates', 'Eurostar Monopoly', 'Delays','Other'}
    return [cat for cat in categories if cat in valid_categories]

def classify_articles_with_ollama(processed_articles, limit=20):
    """
    Classify up to `limit` articles using Ollama's chat-based model with category normalization and counts.
    """
    classified_articles = []
    category_counter = Counter()
    total_articles = min(limit, len(processed_articles))
    
    print(f"[{datetime.now()}] Starting classification of up to {total_articles} articles...\n")
    
    for idx, article in enumerate(processed_articles[:limit], 1):
        title = article['title']
        content = " ".join(article['processed_body'])
        print(f"[{datetime.now()}] Classifying article {idx}/{total_articles}: {title}")
        
        start_time = time.time()
        response = ollama.chat(
            model='llama3.2',
            messages=[
                {
                    'role': 'user',
                    'content': f'{PROMPT} \n\n"{content}"'
                },
            ]
        )
        end_time = time.time()
        
        raw_category = response['message']['content']
        categories = normalize_categories(raw_category)
        category_counter.update(categories)
        
        classified_articles.append({
            "title": title,
            "categories": categories
        })
        print(f"[{datetime.now()}] Completed article {idx} in {end_time - start_time:.2f} seconds.\n")
    
    print(f"[{datetime.now()}] Classification completed for {total_articles} articles.\n")
    print("\nCategory Counts:")
    for category, count in category_counter.most_common():
        print(f"{category}: {count}")
    
    print(f"\nMost Common Category: {category_counter.most_common(1)[0][0]}")
    return classified_articles

classified_articles = classify_articles_with_ollama(processed_articles, limit=20)
for article in classified_articles:
    print(f"Title: {article['title']}")
    print(f"Categories: {', '.join(article['categories'])}")
    print("-" * 50)
```
